# Analyse des résultats de Modèles Linéaraires appliqués au Dataset CIFAR-10

L'objectif du code et de l'apprentissage du modèle linéaire est de vérifier si il est possible de classifier les images du dataset "CIFAR-10". 

###  Importation des données

```python
from tensorflow.keras.datasets import cifar10
```

On charge les données de dans des variables d'entrainement et de validation 

```python
(train_data, train_labels), (val_data, val_labels) = cifar10.load_data()
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

### Initialisation du model

```python
from tensorflow.keras.layers import *
from tensorflow.keras.models import *
```

#### Petit aperçu des données

```python
import numpy as np
import matplotlib.pyplot as plt

samples_per_class =8
def visualize_data(dataset, classes, samples_per_class):
    num_classes = len(classes)
    for y, cls in enumerate(classes):
      idxs = np.flatnonzero(train_labels == y)
      idxs = np.random.choice(idxs, samples_per_class, replace=False)
      for i, idx in enumerate(idxs):
        plt_idx = i * num_classes + y + 1
        plt.subplot(samples_per_class, num_classes, plt_idx)
        plt.imshow(train_data[idx].reshape((32, 32, 3)).astype('uint8'))
        plt.axis('off')
        if i == 0:
          plt.title(cls)
    plt.show()

visualize_data(train_data, classes, samples_per_class)
```

#### Test avec la fonction d'activation softmax

```python
linearModel = Sequential()
linearModel.add(Flatten(input_shape=(32, 32, 3)))
linearModel.add(Dense(10, activation='softmax'))
linearModel.compile(loss='sparse_categorical_crossentropy',
                  metrics=['sparse_categorical_accuracy'])
```

Entrainement du model pour softmax

```python
linearModel.fit(train_data, train_labels, validation_data=(val_data,val_labels), epochs=50, batch_size = 1024)
```

#### Test avec la fonction d'activation relu

```python
linearModel_relu = Sequential()
linearModel_relu.add(Flatten(input_shape=(32, 32, 3)))
linearModel_relu.add(Dense(10, activation='relu'))
linearModel_relu.compile(loss='sparse_categorical_crossentropy',
                  metrics=['sparse_categorical_accuracy'])
```

Entrainement du model pour softmax

```python
linearModel_relu.fit(train_data, train_labels, validation_data=(val_data,val_labels), epochs=50, batch_size = 1024)
```
